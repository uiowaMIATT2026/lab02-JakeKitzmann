```python
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
circle_1 = sitk.ReadImage('data/circle_one.nii.gz')
circle_2 = sitk.ReadImage('data/circle_two.nii.gz')
circle_1_registered = sitk.ReadImage('data/circle_one_registered.nii.gz')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(sitk.GetArrayViewFromImage(circle_1), cmap='gray')
plt.axis('off')
plt.title('Circle One')
plt.subplot(1,2,2)
plt.imshow(sitk.GetArrayViewFromImage(circle_2), cmap='gray')
plt.axis('off')
plt.title('Circle Two')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(sitk.GetArrayViewFromImage(circle_1_registered), cmap='gray')
plt.title('Circle One Registered to Circle Two')
plt.subplot(1,2,2)
plt.imshow(sitk.GetArrayViewFromImage(circle_2), cmap='gray')
plt.title('Circle Two')
```

```python
difference_image = sitk.GetArrayViewFromImage(circle_1_registered) - sitk.GetArrayViewFromImage(circle_2)
plt.imshow(difference_image, cmap='Reds')
plt.title('Difference Image Between fixed and moving images')
plt.colorbar()
```
